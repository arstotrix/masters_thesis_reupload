```python
import os
from datetime import datetime
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def listjoiner(list):
    lister=''
    for a in list:
                lister = lister+', '+a.text
    lister = lister[2:]
    return lister
```

```python
def checktext(li, param):
    try:
        return li.dl.find_all('dd', class_=param)[0].text
    except IndexError:
        return 0
    except AttributeError:
        return 0
```

```python
work_ids=[]
work_titles=[]
authors=[]
fandoms=[]
iscrossover=[]
ratings=[]
warnings=[]
genres=[]
completions=[]
dates=[]
pairings=[]
characters=[]
freeforms=[]
summaries=[]
languages=[]
wordcounts=[]
chaptercounts=[]
commentcounts=[]
kudoscounts=[]
bookmarks=[]
hitcounts=[]
```

```python
for root, dirs, files in os.walk('ao3_data/ao3_data_23'):
    for name in sorted(files):
        fname = os.path.join(root, name)
        print(fname)
        with open(fname, encoding='utf-8') as f:
            tempor = f.read()
        tempor=''.join(tempor.split('""'))[1:-1]
        tempor=tempor.replace('\"','')
        tempor=tempor.replace("\\n",' ')
        tempor=tempor.replace("\\",'')
        soup = BeautifulSoup(tempor)
        lis = soup.find_all('li', role="article")
        for li in lis:
                licrossover=0
                liid = li['id'][5:]
                l4 = li.h4
                try:
                        liname = l4.find_all('a')[0].text
                        liauthor = l4.find_all('a')[1].text
                except IndexError:
                        liname = l4.a.text
                        liauthor = 'Anonymous'
                lifandoms = listjoiner(li.h5.find_all('a'))
                if len(lifandoms.split(', '))>1:
                        licrossover=1
                lirating = li.div.ul.find_all('a')[0].text
                liwarnings = li.div.ul.find_all('a')[1].text
                ligenre = li.div.ul.find_all('a')[2].text
                licomplete = li.div.ul.find_all('a')[3].text
                lidate = datetime.strptime(li.p.text, '%d %b %Y').strftime('%Y/%m/%d')
                lipairings = listjoiner(li.find_all(class_='tags')[0].find_all('li', class_='relationships'))
                lichars = listjoiner(li.find_all(class_='tags')[0].find_all('li', class_='characters'))
                lifrees = listjoiner(li.find_all(class_='tags')[0].find_all('li', class_='freeforms'))
                try:
                        lisummary = li.blockquote.text
                except AttributeError:
                        lisummary = "-"
                lilanguage = checktext(li, 'language')
                liwordcount = checktext(li, 'words')
                lichapters = checktext(li, 'chapters')
                licomment = checktext(li, 'comments')
                likudos = checktext(li, 'kudos')
                libookmark = checktext(li, 'bookmarks')
                lihits= checktext(li, 'hits')
                work_ids.append(liid)
                work_titles.append(liname)
                authors.append(liauthor)
                fandoms.append(lifandoms)
                iscrossover.append(licrossover)
                ratings.append(lirating)
                warnings.append(liwarnings)
                genres.append(ligenre)
                completions.append(licomplete)
                dates.append(lidate)
                pairings.append(lipairings)
                characters.append(lichars)
                freeforms.append(lifrees)
                summaries.append(lisummary)
                languages.append(lilanguage)
                wordcounts.append(liwordcount)
                chaptercounts.append(lichapters)
                commentcounts.append(licomment)
                kudoscounts.append(likudos)
                bookmarks.append(libookmark)
                hitcounts.append(lihits)
                #print(li)
```

```python
df = pd.DataFrame()
df['work_ids'] = work_ids
df['work_titles']=work_titles
df['authors']=authors
df['fandoms']=fandoms
df['iscrossover']=iscrossover
df['ratings']=ratings
df['warnings']=warnings
df['genres']=genres
df['completions']=completions
df['dates']=dates
df['pairings']=pairings
df['characters']=characters
df['freeforms']=freeforms
df['summaries']=summaries
df['languages']=languages
df['wordcounts']=wordcounts
df['chaptercounts']=chaptercounts
df['commentcounts']=commentcounts
df['kudoscounts']=kudoscounts
df['bookmarks']=bookmarks
df['hitcounts']=hitcounts
```
